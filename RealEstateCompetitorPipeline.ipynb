```python
# I am grouping the data by color contrast which creates a lazy groupby dataframe object
# You can confirm this by printing the groups object call in our variable

grouped_data = df.groupby("color_contrast")

# I am now summing this lazy object by the total funding raised, which is a dataframe with columns
# High Contrast and Low Contrast has rows inside of it, which is why the .sum() works!
# {'High Contrast': [0, 1, 2], 'Low Contrast': [3]}

funding_sums = grouped_data['total_funding_raised'].sum()
```

```python
# These are the data points that I think will be good for analysis
# Analyze total/average funding by color contrast/collapsable navbar/color_harmony_normalized_design_strategy
# Analyze market standing by color pairings and design strategy
```

```python
# 1. Create the bar plot (Same as before)
sns.barplot(data=df, x='color_contrast', y='total_funding_raised', estimator='sum', errorbar=None)

# 2. Tell the Y-axis to use Dollar formatting
# This tells the 'current axis' (gca) to format the Y numbers as currency ($)
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))

# 3. Add labels and title
plt.title('Total Funding Raised by Color Contrast')
plt.ylabel('Total Funding (USD)')
plt.xlabel('Color Contrast Type')

plt.show()
```

```python
sns.barplot(data=df, y="total_funding_raised", x="navbar_collapsable", estimator="sum", errorbar=None)

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))


plt.title('Total Funding Raised by Collapsable Navbar UI')
plt.ylabel('Total Funding (USD)')
plt.xlabel('Is navbar collapsable?')

plt.show()
```

```python
plt.figure(figsize=(10, 5)) 

sns.barplot(data=df, y="total_funding_raised", x="color_harmony", estimator="sum", errorbar=None)

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))


plt.title('Total Funding Raised by Color Harmony UI')
plt.ylabel('Total Funding (USD)')
plt.xlabel('Color Harmony Pairings')

plt.show()
```
