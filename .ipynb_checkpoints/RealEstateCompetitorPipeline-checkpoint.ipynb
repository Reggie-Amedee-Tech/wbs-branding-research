```python
# We are currently developing a real estate management application for a client. Being that our client is still finalizing their branding,
# we were tasked with looking at other similar applications in hopes of coming up with a branding revamp. By branding, we mean the
# applications color schemes, UI components, and typography. We will be judging these applications based on 3 factors,
# background color, overall color contrast, and navigation bar layout.
```

```python
# I am creating a company class that will have 4 attributes. name, total funding raised, navbar collapsable, 
# color contrast type, and background
# name: string
# total funding raised: int
# navbar collapsable: boolean
# background: Enum -> nuetral, non-nuetral
# color contrast type: Enum -> high contrast, low contrast
```

```python
from enum import Enum
import pandas as pd
import csv
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
from google.cloud import bigquery
```

```python
client = bigquery.Client()

class ColorContrastType(Enum):
    HighContrast = "High Contrast"
    LowContrast = "Low Contrast"

class Background(Enum):
    Neutral = "Neutral"
    NonNeutral = "Non Neutral"

class ColorHarmony(Enum):
    Monochromatic = "Monochromatic"
    Achromatic = "Achromatic"
    Achromaticn1 = "Achromatic+1"
    HighContrast = "High Contrast"
    Analogous = "Analogous"
    Complementary = "Complementary"

class NormalizedDesignStrategy(Enum):
    TrustLed = "Trust-led"
    ModernMinimal = "Modern Minimal"
    HighAction = "High Action"

class PopularityBucket(Enum):
    MarketLeader = "Market Leader"
    HighGrowth = "High Growth"
    NicheSpecialized = "Niche/Specialized"

class Company:
    def __init__(self, name, total_funding_raised, navbar_collapsable, background, color_contrast, color_harmony, normalized_design_strategy, popularity_bucket):
        self.name = name
        self.total_funding_raised = total_funding_raised
        self.navbar_collapsable = navbar_collapsable
        self.background = background
        self.color_contrast = color_contrast
        self.color_harmony = color_harmony
        self.normalized_design_strategy = normalized_design_strategy
        self.popularity_bucket = popularity_bucket

    def display_info(self):
        return f"Company Name: {self.name},\n Company Total Funding: ${self.total_funding_raised},\n Company Navbar Collapsable: {self.navbar_collapsable},\n Company UI Background: {self.background},\n Company UI Color Contrast: {self.color_contrast} \n Company UI Color Harmony: {self.color_harmony}, \n Company UI Normalized Design Strategy: {self.normalized_design_strategy}, \n Company market standing: {self.popularity_bucket}"
        
```

```python
# filename = "competitor_research_matrix.csv" -> Getting data from google bigquery now!

sqlquery = "SELECT * FROM `real-estate-apps-data-pipeline.realestatedataflow.cleaned_competitor_list`"

# df = pd.read_csv(filename, index_col=False)

df = client.query(sqlquery).to_dataframe()

df.head()
```

```python
# I am grouping the data by color contrast which creates a lazy groupby dataframe object
# You can confirm this by printing the groups object call in our variable

grouped_data = df.groupby("color_contrast")

# I am now summing this lazy object by the total funding raised, which is a dataframe with columns
# High Contrast and Low Contrast has rows inside of it, which is why the .sum() works!
# {'High Contrast': [0, 1, 2], 'Low Contrast': [3]}

funding_sums = grouped_data['total_funding_raised'].sum()

```

```python
# Make sure labels fit 

# 1. Create the bar plot (Same as before)
sns.barplot(data=df, x='color_contrast', y='total_funding_raised', estimator='sum', errorbar=None)

# 2. Tell the Y-axis to use Dollar formatting
# This tells the 'current axis' (gca) to format the Y numbers as currency ($)
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))

# 3. Add labels and title
plt.title('Total Funding Raised by Color Contrast')
plt.ylabel('Total Funding (USD)')
plt.xlabel('Color Contrast Type')

plt.show()
```

```python
sns.barplot(data=df, y="total_funding_raised", x="navbar_collapsable", estimator="sum", errorbar=None)

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))


plt.title('Total Funding Raised by Collapsable Navbar UI')
plt.ylabel('Total Funding (USD)')
plt.xlabel('Is navbar collapsable?')

plt.show()
```

```python
sns.barplot(data=df, y="total_funding_raised", x="color_harmony", estimator="sum", errorbar=None)

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))


plt.title('Total Funding Raised by Color Harmony UI')
plt.ylabel('Total Funding (USD)')
plt.xlabel('Is navbar collapsable?')

plt.show()
```

```python
sns.barplot(data=df, y="total_funding_raised", x="color_contrast", estimator="mean", errorbar=None)

plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))


plt.title('Average Funding Raised by Color contrast')
plt.ylabel('Avg Funding (USD)')
plt.xlabel('High vs Low contrast')

plt.show()
```
